```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100) 

y = 2 * x**2 + 3 * x + 5

plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label=f'y = 2x^2 + 3x + 5')  
plt.xlabel('Значения x')  
plt.ylabel('Значения y')  
plt.title('График квадратичной функции')  
plt.legend()  

plt.grid(True)  
plt.show()
```

```python
A = 3.0 
B = 1.7  
C = np.pi / 3

x = np.linspace(0, 2 * np.pi, 100)
y = A * np.sin(B * x + C)

plt.figure(figsize=(8, 6))  
plt.plot(x, y, label=f'y = 1 * sin(2 * x + {C:.2f})', color='b')  
plt.xlabel('Значения x')  
plt.ylabel('Значения y')  
plt.title('График гармонической функции')  
plt.legend()  


plt.grid(True) 
plt.show()
```

```python
plt.figure(figsize=(12, 6))

y_harmonic = A * np.sin(B * x + C)

plt.subplot(1, 2, 1) 
plt.plot(x, y_harmonic, label=f'y = {A} * sin({B} * x + {C:.2f})', color='b')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.title('Гармоническая функция')
plt.legend()
plt.grid(True)

a = 2
b = 4
c = 3

x_quad = np.linspace(-10, 10, 100) 
y_quadratic = a * x_quad**2 + b * x_quad + c

plt.subplot(1, 2, 2) 
plt.plot(x_quad, y_quadratic, label=f'y = {a}x^2 + {b}x + {c}', color='r')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.title('Квадратичная функция')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x, y_harmonic, label=f'y = {A} * sin({B} * x + {C:.2f})', color='b')
plt.plot(x, y_quadratic, label=f'y = {a}x^2 + {b}x + {c}', color='r')

plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.title('Графики гармонической и квадратичной функций')
plt.legend(loc='upper right') 

plt.grid(True)
plt.show()
```

```python
t = np.linspace(0, 4 * np.pi, 1000)

x = np.cos(3 * t)
y = np.sin(2 * t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='x(t) = cos(3t), y = sin(2t)')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.title('График параметрической функции')
plt.legend()

plt.grid(True)
plt.show()
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
min_income = [
    4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32
]

plt.figure(figsize=(10, 6))
plt.bar(years, min_income, color='blue', alpha=0.5)
plt.xlabel('Годы')
plt.ylabel('Прожиточный минимум')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.xticks(years) 

plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(min_income, labels=years, autopct='%1.1f%%', startangle=140)
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')

plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2  

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Каркасная поверхность')

plt.show()
```
